```python
# importação de bibliotecas

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# vendo informações do dataframe

df.info()
```

```python
# lista de um elemento contendo a string purpose

obtem_purpose = ['purpose']
```

```python
# definindo novo dataframe, com a nova coluna

final_data = pd.get_dummies(df, columns=obtem_purpose, drop_first=True)
```

```python
# vendo informações do novo dataframe

final_data.info()
```

```python
# importação do treino e teste

from sklearn.model_selection import train_test_split
```

```python
# eliminando a coluna de teste

X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']
```

```python
# splitando o treino e teste

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
# importando a arvore de decisão

from sklearn.tree import DecisionTreeClassifier
```

```python
# inicializando um objeto da classe

dtree = DecisionTreeClassifier()
```

```python
# fitando o modelo

dtree.fit(X_train, y_train)
```

```python
# tirando predições do modelo

pred = dtree.predict(X_test)
```

```python
# importar a classificação do modelo

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# avaliar a precisão do modelo

print(classification_report(y_test, pred))
```

```python
# matriz de confusão

print(confusion_matrix(y_test, pred))
```

```python
# importando a floresta aleatória

from sklearn.ensemble import RandomForestClassifier
```

```python
# inicializando um objeto da classe

rfc = RandomForestClassifier(n_estimators=500)
```

```python
# fitando o modelo

rfc.fit(X_train, y_train)
```

```python
# tirando predições do modelo

rfc_pred = rfc.predict(X_test)
```

```python
# avaliar a precisão do modelo

print(classification_report(y_test, rfc_pred))
```

```python
# matriz de confusão

print(confusion_matrix(y_test, rfc_pred))
```
